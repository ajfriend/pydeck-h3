```python
import h3
import pydeck_h3 as pdh
```

```python
d = {'type': 'Polygon',
 'coordinates': [[[-122.38902568817139, 37.76871536203574],
   [-122.38899081945418, 37.76847153193986],
   [-122.38891571760176, 37.768255264573305],
   [-122.38826930522917, 37.76828070782571],
   [-122.38820761442184, 37.7682997902593],
   [-122.38804399967194, 37.768452449550594],
   [-122.38779455423354, 37.76859238695745],
   [-122.38781332969666, 37.76880017231546],
   [-122.38815933465959, 37.76877684949823],
   [-122.3882183432579, 37.768798052059644],
   [-122.3887601494789, 37.768768368471946],
   [-122.38878160715103, 37.768747165902056],
   [-122.38902568817139, 37.76871536203574]]]}
```

```python
out = h3.polyfill(d, 15, geo_json_conformant=True)
pdh.plot(out, line_width=.1)
```

```python
h3.cell_area(list(out)[0], unit='m^2')
```

```python
out2 = h3.compact(out)
pdh.plot(out2, line_width=.1)
```
