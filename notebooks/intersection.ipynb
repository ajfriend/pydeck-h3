```python
def rich_compare(a, b):
    if a == b:
        return 0
    
    res_a = h3.h3_get_resolution(a)
    res_b = h3.h3_get_resolution(b)

    if res_a > res_b:
        if h3.h3_to_parent(a, res_b) == b:
            return -1
        
    if res_a < res_b:
        if h3.h3_to_parent(b, res_a) == a:
            return +1
    
    mask = 0x800f_ffff_ffff_ffff
    
    if (a & mask) < (b & mask):
        return -2
    else:
        return +2
    
def hex_key(h):
    mask = 0x800f_ffff_ffff_ffff
    return h3.string_to_h3(h) & mask
```
