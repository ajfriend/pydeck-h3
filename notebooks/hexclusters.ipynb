```python
import requests

from shapely import wkt
import shapely
import shapely.geometry

import h3
import pydeck_h3 as pdh
```

```python
url = 'https://data.sfgov.org/api/views/wamw-vt4s/rows.json?accessType=DOWNLOAD'
r = requests.get(url, allow_redirects=True)
d = r.json()
d = {
    row[-1]: shapely.geometry.mapping(wkt.loads(row[-4]))
    for row in d['data']
}
```

```python
list(d.keys())
```

```python
def foo(d):
    for v in d.values():
        for poly in v['coordinates']:
            g3 = {
                'type': 'Polygon',
                'coordinates': poly,
            }
            hexset = h3.polyfill(g3, 8, geo_json_conformant=True)
            if len(hexset) > 30:
                yield hexset
                
sets = list(foo(d))
list(enumerate(map(len, sets)))
```

```python
pdh.plot(sets[5]).to_html()
```
