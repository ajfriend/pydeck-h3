```python
import util2 as u
import h3
import toolz

h = '89283082807ffff'

h = h3.geo_to_h3(37.787598, -122.409969, 9)

h2 = h3.geo_to_h3(37.725531, -122.461333, 9)
```

```python
def ring_weights(K, i):
    """Smoothing weights for ring i in disk K
    0 <= i <= K
    The sum of the weights of all hexes in the ring is 1.
    """
    return (K + 1.0 - i)/(K+1)**3
```

```python
ring_weights(10, 1)
```

```python
K = 6

hexes = h3.k_ring(h, K)

hexes = {
    n: ring_weights(K, h3.h3_distance(h, n))
    for n in hexes
}
```

```python
K = 8

hexes2 = h3.k_ring(h2, K)

hexes2 = {
    n: ring_weights(K, h3.h3_distance(h2, n))
    for n in hexes2
}

```

```python
hexes3 = toolz.merge(hexes, hexes2)
```

```python
u.plot_hexvals3D(hexes3, opacity=.4, elevation_scale=100000).to_html()
```
